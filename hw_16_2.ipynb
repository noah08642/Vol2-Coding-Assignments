## Problem 16.10

Code up the value iteration method for finding the optimal value function $V(0,T,s)$ in the $3 \times 3$ grid motion problem. Your code should accept a time $T$ and a dictionary $u$ of rewards for each action and return the value function $V(0,T,s)$ for every $s \in \{1,2,\ldots,9\}$.

```python

def grid_value_iteration(T: int, u: dict[int, float]) -> list[float]:
    """Value iteration on the 3x3 grid motion problem.
    
    Parameters:
        T (int): The final time (grid motion problem 
            iterates from [0,T])
        u (dict[int,float]): A dictionary of rewards for
            stepping onto each grid square. Keys are grid
            square indexes 1, 2, ..., 9, and values are the
            reward for stepping onto that square.
    
    Returns:
        list[float]: 9-item list, where each value is V(0,T,s)
            for every s in {1,2,...,9}. The values should be 
            ordered for the different grid indexes from left-to-right,
            top-to-bottom.
    """
    # Allowed state transitions from textbook 3x3 grid problem
    # (key is starting state, value is list of possible next states)
    ALLOWED_TRANSITIONS = {
        1: [2,4],
        2: [1,3,5],
        3: [2,6],
        4: [1,5,7],
        5: [2,4,6,8],
        6: [3,5,9],
        7: [4,8],
        8: [5,7,9],
        9: []
    }
    SQUARES = range(1, len(u) + 1)

    prev, curr = {}, {} # prev here is V(0, 0, s)
    for i in SQUARES: # populate for keys 1, 2, ..., 9
        if not ALLOWED_TRANSITIONS[i]:  # terminal state
            prev[i] = 0
        else:
            prev[i] = max(u[j] for j in ALLOWED_TRANSITIONS[i])


    for t in range(1, T + 1): # iterate until T (inclusive)
        for i in SQUARES: #(go through each square)
            if not ALLOWED_TRANSITIONS[i]:  # terminal state
                curr[i] = 0
            else:
                curr[i] = max(u[j] + prev[j] for j in ALLOWED_TRANSITIONS[i]) # bellman equation

        prev = curr.copy()

    # reformat dict to list

    return [prev[i] for i in SQUARES]
```
