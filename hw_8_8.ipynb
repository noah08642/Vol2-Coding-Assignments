### Part (i)

Code up a method `sample_f()` that takes a function $f: [0,1) \to \mathbb{R}$ and an integer $n > 0$, and returns an array of the values of the function $f$ at the points $\frac{k}{2^n}$ for $k \in \{0, \ldots, 2^n -1\}$.

```python
def sample_f(
    f: Callable[[float | np.ndarray], float | np.ndarray],
    n: int
) -> np.ndarray:
    """Samples the function f at 2^n points from 0 to (2^n - 1)/(2^n)"""
    k = np.arange(0, 2**n)
    x = k / (2**n)
    return f(x)
```

### Part (ii)

Write a method `wavelet_func()` which accepts an array $\mathbf{a} = [a_0, a_1, \ldots, a_{2^n -1}]$ and a value $x \in [0,1)$, and returns $\sum_{k=0}^{2^n - 1}a_k \rho_{n,k}(x)$

```python
def wavelet_func(a: np.ndarray, x: float | np.floating) -> float | np.floating:
    """Given coefficients, returns the son-wavelet summation at a given point."""
    
    n = len(a)
    k = np.floor(n * x).astype(int)

    return a[k]
        
```

### Part (iii)

For the function 

$$f(t) = \frac{\sin(2\pi t - 5)}{\sqrt{|t - \pi / 20|}}$$

and for each of the values $n \in \{1, 2, 3, \ldots, 10\}$, use your code to plot:

1. $f(t)$, and
2. $f_n(t) = \sum_{k=0}^{2^n - 1} f(k/2^n) \rho_{n,k}(x)$

Make sure to label which curve is which, and to title your plots, etc., before displaying them to the screen.

```python
f = lambda x: np.sin(2 * np.pi * x - 5) / np.sqrt(np.abs(x - np.pi / 20))
x = np.linspace(0, 1, 1000, endpoint=False)
f_true = f(x)

for n in range(1, 11):
    a = sample_f(f, n)
    plt.plot(x, f_true, label='f(t)')
    plt.plot(x, wavelet_func(a, x), label=f'f_n(t)')
    plt.title(f'n = {n}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
