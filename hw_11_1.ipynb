## Problem 11.4 

Consider the series $\displaystyle \sum_{n=1}^{10^7} \frac{1000}{n}$

### Part (i)

In the following cell, sum the series forward from 1 to $10^7$ using a for-loop, and display the result to the screen.

```python
s = 0
for n in range(1, 10**7 + 1):
    s += 1000 / n 

print(s)
```

Now, in the next cell, sum the series backward from $10^7$ down to 1 using a for-loop, and display the result to the screen.

```python
s = 0
for n in range(10**7, 0, -1):
    s += 1000 / n 

print(s)
```

### Part (ii)

Floating-point addition is commutative, so why does changing the order of this summation change the result?

**RESPONSE:** _The order changes how each result snaps into the nearest float, and those changes add up over time.  In other words floating point operations do not necessary hold the associative property of addition_

## Problem 11.5

Plot the function
$$f(x) = (1-x) - 1$$
and the function
$$g(x) = x$$
at 1000 equally-spaced points along the interval $[-3 \times 10^{-15}, +3 \times 10^{-15}]$.

(Note that $f(x) = -x$ in exact arithmetic)

```python
x_pts = np.linspace(-3e-15, 3e-15, 1000)    # 1000 equally-spaced points

plt.plot(x_pts, (np.ones_like(x_pts) - x_pts) - np.ones_like(x_pts), label="f(x)")
plt.plot(x_pts, x_pts, label="g(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("comparison of f(x) and g(x)")
```

Explain why the graph of $x$ looks smooth but the graph of $(1-x) - 1$ is jagged. 

**RESPONSE:** _Because the subtraction operations in the parentheses and then subsequent subtraction outside of the parentheses created small snapping erros which were then added or subtracted from to or from the final value_

Now, plot the ratio of $\displaystyle \frac{f(x)}{g(x)} = \frac{(1-x) - 1}{x}$. 

(Notice that, in exact arithmetic, $\displaystyle \frac{f(x)}{g(x)} = \frac{-x}{x} = -1$.) 

```python
x_pts = np.linspace(-3e-15, 3e-15, 1000)    # 1000 equally-spaced points

plt.plot(x_pts, ((np.ones_like(x_pts) - x_pts) - np.ones_like(x_pts)) / x_pts)
```

Why does the error (the distance from the correct value of $-1$) get larger as $x \to 0$?

**RESPONSE:** _Think of (1 - x) - 1 as -x + delta where delta is the error.  When you divide both sides by x, we get f/g = -1 + delta/x, which explodes as x nears zero_
