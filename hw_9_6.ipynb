```python
import numpy as np
```

```python
n = np.array([4, 16, 64, 256, 1024, 4096])
trap_err = np.array([1.3803e-00, 0.1074e-00, 6.9562e-03, 4.3600e-04, 2.7255e-05, 1.7034e-06])
simp_err = np.array([0.3914e-00, 0.0100e-00, 7.7541e-05, 3.3449e-07, 1.3162e-09, 5.1585e-12])

log_n = np.log(n)
log_trap = np.log(trap_err)
log_simp = np.log(simp_err)

# Linear regression to get slope a
a_trap, log_k_trap = np.polyfit(log_n, log_trap, 1)
a_simp, log_k_simp = np.polyfit(log_n, log_simp, 1)

print("Estimated a for trapezoidal rule:", a_trap)
print("Estimated a for Simpson's rule:", a_simp)
```
