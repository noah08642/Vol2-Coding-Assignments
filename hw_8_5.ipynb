## Problem 8.27(ii)

The leakage amplitude for the DFT of a noncharacteristic sinusoidal frequency is given by the formula
$$|\hat{f}(k)| = \left| \frac{\sin [\pi(\lambda - k)]}{N \sin \left(\frac{\pi(\lambda - k)}{N}\right)} \right|$$

Fix $N = 100$. 

Write a function `leakage_amplitude` that returns $|\hat{f}(k)|$ as a function of $k$ and $\lambda$.

Your function should accept a numpy array `k` of values to evaluate the function at, and a floating-point number `lam` (representing $\lambda$) and should return a numpy array $|\hat{f}$(`k`)$|$ which matches the size of `k`.



```python
def leakage_amplitude(k: np.ndarray, lam: float) -> np.ndarray:
    """Returns the leakage amplitude at k and lambda (assuming N=100)"""
    
    N = 100

    top = np.sin(np.pi*(lam - k))
    bottom = N * np.sin(np.pi * (lam - k) / N)

    return np.abs(top / bottom)
```

Now, in the following code cells, create the following plots (with titles, axis labels, etc.) and display them:

1. $|\hat{f}(k)|$ versus $k$, for $k \in [0, 99]$ and $\lambda  = 1.1$

2. $|\hat{f}(k)|$ versus $k$, for $k \in [0, 99]$ and $\lambda  = 10.4$

3. $|\hat{f}(k)|$ versus $k$, for $k \in [0, 99]$ and $\lambda  = 50.2$
   
4. $|\hat{f}(k)|$ versus $k$, for $k \in [0, 99]$ and $\lambda  = 89.5$

```python
k = np.arange(0, 100, 1)
lam = 1.1
y = leakage_amplitude(k, lam)
plt.plot(k, y)
plt.xlabel("k")
plt.ylabel("abs(f_hat(k))")
plt.title(f"abs(f_hat(k)) vs k for k in [0,99] and lam = {lam}")
```
