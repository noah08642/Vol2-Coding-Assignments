## Problem 8.32

Let $f(x) = x$ and $g(x) = \sin(x)$. 

First, fill in the functions `sample_f` and `sample_g`, which accept as an argument `n` (the number of samples to use), and should return numpy arrays of length `n` containing equally-spaced samples of $f$ and $g$ (respectively) on the interval $[0, 2\pi]$

```python
def sample_f(n: int) -> np.ndarray:
    """Samples the function f(x) = x at n equally-spaced points on [0, 2pi].
    
    Args:
        n (int) - The number of samples to generate
        
    Returns:
        np.ndarray - The resulting array of n samples of f(x),
            taken at n equally-spaced points
    """
    
    return np.linspace(0, 2*np.pi, n)
```

```python
def sample_g(n: int) -> np.ndarray:
    """Samples the function g(x) = sin(x) at n equally-spaced points on [0, 2pi].
    
    Args:
        n (int) - The number of samples to generate
        
    Returns:
        np.ndarray - The resulting array of n samples of g(x),
            taken at n equally-spaced points
    """
    return np.sin(np.linspace(0, 2*np.pi, n))
```

Now, using these functions, generate two arrays `f` and `g`, both of which are 1000 equally-spaced samples of the functions $f(x)$ and $g(x)$, respectively, on $[0, 2\pi]$

```python
n = 1000

f = sample_f(n)
g = sample_g(n)
```

Next, fill in the functions `conv` and `hadamard`, which accept as arguments `f` and `g` (two equally-sized numpy arrays), and return (i) $f * g$ (the circular convolution of $f$ and $g$), and (ii) $f \odot g$ (the Hadamard product of $f$ and $g$), respectively, as numpy arrays.

```python
def conv(f: np.ndarray, g: np.ndarray) -> np.ndarray:
    """Computes the circular convolution of f and g.
    
    Args:
        f (np.ndarray) - Array of samples 
        g (np.ndarray) - Array of samples 
    
    Returns:
        np.ndarray - The circular convolution of f and g
    """
    n = max(len(f), len(g))
    return np.fft.ifft(np.fft.fft(f, n) * np.fft.fft(g, n)).real
```

```python
def hadamard(f: np.ndarray, g: np.ndarray) -> np.ndarray:
    """Computes the Hadamard product of f and g.
    
    Args:
        f (np.ndarray) - Array of samples 
        g (np.ndarray) - Array of samples 
    
    Returns:
        np.ndarray - The Hadamard product of f and g
    """
    return np.multiply(f, g)
```

Now, in the following code cell, _on different plots_, plot a line plot connecting the points $(x_k, y_k)$ where $x_k = \frac{2\pi k}{1000}$, and $y_k$ takes on the following values:

1. `f` (the samples from $f(x)$ you calculated from earlier)
2. `g` (the samples from $g(x)$ you calculated from earlier)
3. $f * g$ (the circular convolution of `f` and `g`), and
4. $f \odot g$ (the Hadamard product of `f` and `g`)

Make sure each of your plots has a title, and is displayed to the screen.

```python
f_star_g = conv(f, g)
f_dot_g = hadamard(f, g)


plt.plot(f)
plt.title("f(x) = f")
plt.xlabel("x")
plt.ylabel("y")

plt.figure()
plt.plot(g)
plt.title("g(x) = sin(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.figure()
plt.plot(f_star_g)
plt.title("Circular Convolution")
plt.xlabel("x")
plt.ylabel("(f * g)(x)")

plt.figure()
plt.plot(f_dot_g)
plt.title("Hadamard Product")
plt.xlabel("x")
plt.ylabel("(f o. g)(x)")
```
